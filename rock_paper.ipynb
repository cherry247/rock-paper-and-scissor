```
import ipywidgets as widgets
import random
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import Layout, Button, Box, Text, Textarea, Dropdown, Label, IntSlider
from IPython.display import display
```

```
def analysis(df):
  form_item_layout = Layout(
    display='flex',
    flex_flow='row',
    justify_content='space-between'
  )

  form_items = [
    Box([Label(value='Total matches'),
         Text(value='5')], layout=form_item_layout),
   
    Box([Label(value='User win'),
         Text(value='{}'.format(df['u'].sum()))], layout=form_item_layout),
    Box([Label(value='Computer win'),
         Text(value='{}'.format(df['c'].sum()))], layout=form_item_layout),
    Box([Label(value='Tie matches'),
         Text(value='{}'.format(df['d'].sum()))], layout=form_item_layout)

   
  ]

  form = Box(form_items, layout=Layout(
    display='flex',
    flex_flow='column',
    border='solid 2px',
    align_items='stretch',
    width='50%',
    color='white'
  ))
  return form
```

```
print('enter the number of time you want to play the game')
w= widgets.IntSlider()
display(w)
```

```
n=w.value
computer=['rock','paper','scissor']
u=[]
c=[]
d=[]
for i in range(n):
  user=input('r,p,s')  #we will use button here
  comp=random.choice(computer)
  if user=='rock' and  comp=='rock':
    print("no one gets point")
    u.append(0)
    c.append(0)
    d.append(1)
  elif user=='rock' and  comp=='scissor':
    print('booya! you  get a point')
    u.append(1)
    c.append(0)
    d.append(0)
  elif user=='rock' and  comp=='paper':
    print('sorry! computer gets point')
    u.append(0)
    c.append(1)
    d.append(0)
  elif user=='paper' and  comp=='scissor':
    print('sorry! computer gets point')
    u.append(0)
    c.append(1)
    d.append(0)
  elif user=='paper' and  comp=='rock':
    print('booya! you  get a point')
    u.append(1)
    c.append(0)
    d.append(0)
  elif user=='paper' and  comp=='paper':
    print("no one gets point")
    u.append(0)
    c.append(0)
    d.append(1)
  elif user=='scissor' and  comp=='scissor':
    print("no one gets point")
    u.append(0)
    c.append(0)
    d.append(1)
  elif user=='scissor' and  comp=='paper':
    print('booya! you  get a point')
    u.append(1)
    c.append(0)
    d.append(0)
  elif user=='scissor' and  comp=='rock':
    print('sorry! computer gets point')
    u.append(0)
    c.append(1)
    d.append(0)
  
data=pd.DataFrame({'u':u,'c':c,'d':d})
if data['u'].sum() > data['c'].sum():
  print('you win')
elif data['c'].sum() > data['u'].sum():
  print('you lose,computer wins')
else:
  print('its Tie')
f=analysis(data)
f
```
